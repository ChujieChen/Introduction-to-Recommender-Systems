```python
import numpy as np
import pandas as pd
```

```python
docs = pd.read_csv("HW2-docs.csv")
users = pd.read_csv("HW2-users.csv")
docs = docs.iloc[:,1:]
users = users.iloc[:,1:]
docs.index = range(1,len(docs)+1)
users.index = range(1,len(users)+1)
```

```python
docs
```

```python
users
```

# Part 1. Build and use a very basic profile

```python
basic_user = (docs.T).dot(users.fillna(0))
basic_user.head()
```

```python
basic_pred = docs.dot(basic_user)
basic_pred.T
```

```python
basic_pred.iloc[:,0].sort_values(axis=0, ascending=False).head()
```

```python
(basic_pred.iloc[:,1] < 0).sum()
```

# Part 2. Next, let’s treat all articles as having unit weight ...

```python
unit_docs = docs.copy()
unit_docs = unit_docs.div(np.sqrt(unit_docs.sum(axis=1)), axis=0)
```

```python
unit_docs.head()
```

```python
unit_users = (unit_docs.T).dot(users.fillna(0))
unit_users.head()
```

```python
unit_pred = unit_docs.dot(unit_users)
unit_pred.T
```

```python
unit_pred.iloc[:,0].sort_values(axis=0, ascending=False).head()
```

# Part 3. Finally, let’s consider how common different terms are among our documents …

DF is the number of documents in which each content attribute occurs

```python
DF = (unit_docs > 0).sum(axis=0)
DF
```

IDF (inverse document frequency) (1/log(DF) is usually used with large datasets, but we don't need the log here)

```python
IDF = 1/DF
IDF
```

document vector * profile * IDF
    mxn            nx2       n

```python
unit_docs.shape, unit_users.shape, IDF.shape
```

```python
tfidf_pred = unit_docs.multiply(IDF).dot(unit_users)
tfidf_pred.T
```
